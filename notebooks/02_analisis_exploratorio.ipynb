# Análisis de Siniestros Viales en Guadalajara y Zapopan (2019–2025)
## Notebook 02: Análisis Exploratorio

En este notebook exploramos a fondo la base de datos de accidentes viales en Guadalajara y Zapopan. Nuestro objetivo es detectar patrones clave para entender el contexto actual y generar recomendaciones que puedan guiar decisiones en materia de movilidad, infraestructura vial y seguridad pública.

Analizaremos variables como:
- Año y evolución temporal
- Municipios con mayor número de siniestros
- Tipo de accidente
- Gravedad de las consecuencias
- Hora y día de ocurrencia
- Tipo de usuario afectado

Todos los análisis están acompañados de visualizaciones para facilitar su interpretación.


```python
import matplotlib.pyplot as plt
import pandas as pd

gdf = pd.read_csv("../data/accidentes_cleaned.csv")

# Configuración general
plt.style.use("seaborn-v0_8-whitegrid")
plt.rcParams["figure.figsize"] = (10, 5)

# Asegurar orden correcto para días
dias_orden = ["lunes", "martes", "miércoles", "jueves", "viernes", "sábado", "domingo"]
gdf["dia_sem"] = gdf["dia_sem"].str.lower()

### 1. Accidentes por año
gdf["anio"].value_counts().sort_index().plot(kind="bar")
plt.title("Accidentes por Año")
plt.xlabel("Año")
plt.ylabel("Cantidad de accidentes")
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()

### 2. Accidentes por municipio
gdf["mun"].value_counts().plot(kind="bar", color="orange")
plt.title("Accidentes por Municipio")
plt.xlabel("Municipio")
plt.ylabel("Cantidad")
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()

### 3. Tipos de siniestro más comunes
gdf["tipo_siniestro"].value_counts().head(10).plot(kind="barh", color="teal")
plt.title("Tipos de Siniestro más Comunes")
plt.xlabel("Cantidad")
plt.ylabel("Tipo de siniestro")
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()

### 4. Consecuencias del accidente
gdf["consecuencia"].value_counts().plot(kind="bar", color="crimson")
plt.title("Consecuencias de los Siniestros")
plt.xlabel("Consecuencia")
plt.ylabel("Cantidad")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

### 5. Accidentes por franja horaria
gdf["rango_hora"].value_counts().sort_index().plot(kind="bar", color="darkblue")
plt.title("Accidentes por Franja Horaria")
plt.xlabel("Hora del día")
plt.ylabel("Cantidad")
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()

### 6. Accidentes por día de la semana
gdf["dia_sem"].value_counts().reindex(dias_orden).plot(kind="bar", color="seagreen")
plt.title("Accidentes por Día de la Semana")
plt.xlabel("Día")
plt.ylabel("Cantidad")
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()

### 7. Tipo de usuario afectado
gdf["tipo_usuario"].value_counts().head(10).plot(kind="barh", color="purple")
plt.title("Tipo de Usuario Involucrado (Top 10)")
plt.xlabel("Cantidad")
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()

### 8. Sexo de los usuarios involucrados
gdf["sexo"].value_counts().plot(kind="bar", color="orchid")
plt.title("Sexo de las Personas Involucradas")
plt.xlabel("Sexo")
plt.ylabel("Cantidad")
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()

### 9. Rango de edad de los usuarios
gdf["rango_edad"].value_counts().sort_index().plot(kind="bar", color="steelblue")
plt.title("Rango de Edad de las Personas Involucradas")
plt.xlabel("Edad")
plt.ylabel("Cantidad")
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

## Interpretación de resultados

A continuación se resumen los principales hallazgos derivados del análisis exploratorio de los datos de siniestros viales en Guadalajara y Zapopan (2019–2025):

### Evolución temporal

- El año 2019 registra la mayor cantidad de accidentes, seguido por una caída pronunciada en 2020, posiblemente relacionada con las restricciones por la pandemia de COVID-19.
- A partir de 2021, los valores se mantienen relativamente estables, con un promedio de alrededor de 1,800 a 2,000 accidentes por año.

### Comparativa entre municipios

- Guadalajara presenta una mayor concentración de siniestros que Zapopan, con una diferencia significativa que podría explicarse por su mayor densidad poblacional, infraestructura vial más compleja y niveles de tráfico más altos.

### Tipo de siniestro

- La mayoría de los incidentes corresponde a colisiones entre dos o más vehículos.
- Le siguen los atropellamientos y colisiones con objetos fijos, lo que indica una clara prioridad en abordar la seguridad tanto para automovilistas como para peatones.

### Consecuencias

- Aunque la mayoría de los accidentes termina con personas ilesas o lesionadas, se reporta una cantidad considerable de fallecimientos.
- Este dato es especialmente importante para definir zonas de atención prioritaria o tipos de accidentes que requieren intervención más agresiva.

### Horarios de mayor riesgo

- Se observa un pico de accidentes entre las 07:00 y 09:00 horas (horario de entrada laboral/escolar) y otro entre las 14:00 y 19:00 horas (salida laboral y mayor actividad en la vía pública).
- Las horas de la madrugada presentan menos incidentes, pero eso no necesariamente implica menor peligrosidad, ya que muchos accidentes fatales ocurren en esos horarios.

### Días de la semana

- Los accidentes se distribuyen relativamente parejo entre lunes y domingo, pero los fines de semana (sábado y domingo) presentan los valores más altos.
- Esto puede asociarse con actividades recreativas, eventos sociales o consumo de alcohol.

### Tipos de usuarios involucrados

- Los vehículos particulares y motociclistas son los más comúnmente involucrados.
- También hay una presencia notable de peatones, ciclistas y usuarios del transporte público, lo que sugiere una necesidad de infraestructura más segura para todos los tipos de movilidad.

### Sexo y edad

- Los hombres están involucrados en una proporción mucho mayor que las mujeres.
- El rango de edad con mayor participación es entre los 18 y 27 años, seguido de los 28 a 37. Este grupo representa una población joven-adulta que podría beneficiarse de campañas de concientización, revisiones de licencia o educación vial.
